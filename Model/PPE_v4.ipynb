```python
from ultralytics import YOLO

# Define paths
data_path = '/content/drive/MyDrive/PEE-Dataset/data.yaml'  # Update this to your data.yaml path

# Load the YOLOv8 model (e.g., YOLOv8n for Nano version)
model = YOLO('yolov8n.pt')  # You can choose other versions like 'yolov8s.pt', 'yolov8m.pt', etc.

# Train the model
model.train(data=data_path, epochs=50, imgsz=640, batch=8, name='yolov8_training')
```

# Evaluate the model
metrics = model.val(data=data_path)
print(metrics)

```python
results = model('/content/drive/MyDrive/construction_worker_laborer.jpg')
```

```python
# Perform inference
results.save(save_dir='/content/drive/MyDrive/PPE-Dataset')  # Save predictions
```
